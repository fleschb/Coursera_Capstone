<h4>
Importing all the required libraries:

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
print("Libraries Imported")
```

<h4>
Importing the text from the website and putting it in a dataframe:

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
rows = soup.find_all('tr')
list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
df = pd.DataFrame(list_rows)
df.head()
```

<h4>
Splitting up the information and cleaning it up:

```python
df1 = df[0].str.split(',', expand=True)
df1[0] = df1[0].str.strip('[')
for i in range (30):
    df1[i] = df1[i].str.strip('\n')
    df1[i] = df1[i].str.strip('\n]')
df2 = df1[df1[1] != " Not assigned"]
df2 = df2.drop([0],axis=0)
df2 = df2[:-4]
df2 = df2.reset_index(drop=True)
df2.head()
```

<h4>
Getting the dataframe in the required format:

```python
df2['Neighborhood'] = ""
for i in range (31):
    for j in range (103):
        if i > 1 and df2[i][j] != None:
            if df2['Neighborhood'][j]=="":
                df2['Neighborhood'][j]= df2[i][j]
            else:
                df2['Neighborhood'][j]=df2['Neighborhood'][j]+ ', ' + df2[i][j]
df2.drop(df2.iloc[:, 2:31], inplace = True, axis = 1)
df2.columns = ['PostalCode','Borough','Neighborhood']
df2.head()
```
